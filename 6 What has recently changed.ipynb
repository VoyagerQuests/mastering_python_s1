#### Built-in Generics
Before

```python
from typing import List, Dict
def f(xs: List[int]) -> Dict[str, int]: ...
```

Now (Python 3.9)

```python
def f(xs: list[int]) -> dict[str, int]: ...
```

#### Union Types
Before

```python
from typing import Union, Optional
x: Union[int, str]
y: Optional[int]
```

Now (Python 3.10)

```python
x: int | str
y: int | None
```

#### Type parameters become syntax
Before

```python
from typing import TypeVar
T = TypeVar("T")

def first(xs: list[T]) -> T: ...
```

Now (Python 3.12)

```python
def first[T](xs: list[T]) -> T: ...
```

#### Type aliases beome statements
Before

```python
from typing import TypeAlias
UserId: TypeAlias = int
```

Now (Python 3.12)

```python
type UserId = int
type Box[T] = dict[str, T]
```

#### Forward references in Annotations
Before (Pre-Python 3.7)

```python
class User:
    def manager(self) -> "User":
        ...
```

Before (Python 3.7 - 3.13)

```python
from __future__ import annotations

class User:
    def manager(self) -> User:
        ...
```

Now (Python 3.14)

```python
class User:
    def manager(self) -> User:
        ...
```
